## Solution 1) Sorting then Finding the consecutive Sequence -- O(Nlon(N))

Sort the list then by doing this all elements will come consecutively then we can check if current element is 1 more than the previuos element or not, if it is then increment the len variable and compare with MaxLen variable, if the current element is equal to previous element then simply move on to the next element, otherwise if the elements are neither consutive nor they are equal then make len=1 and start findince new sequence.

## Solution 2) Finding the starting elements of consecutive sequences and then finding the maximum lengths  -- TC : O(N) , SC : O(N)

For finding out the starting elements of a consecutive sequence we will check whether an element which is 1 less than current element exists in the list or not, if there exists such an element then the current element is not the starting point of a sequence otherwise it is starting Head. Now to check whether (current element-1) exist in list in O(1) time we can store all these elements in a dictionary.
