## Question

## Solution 1 : Brute Force - O(N^2)

Go on Every Element then check whether you find the same element int the array again or not.

## Solution 2: Sorting - O(Nlog(n))

Sort the array, then traverse the array. Since after sorting, all duplicate elements should appear consecutively. So we can check whether current element is equal to next element or not. If current element is equal to next element then these are duplicates, otherwise the same element can't be found again later in the array because of sorting. 

## Solution 3 : Hashmap/Hashset usage - TC : O(N) , SC - O(N)

Create an Empty map, then traverse the array, if the current element is not found in the map then insert this element in the map. But if it is found in the map, then it means the current element was already found in the array and it has appeared again, thus there is duplicacy.
