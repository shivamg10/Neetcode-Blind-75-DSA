## Solution 1) Brute force - O(N^2)

Here We will check for every element is there any other element present in the array which can pair up with the current element to achieve the target.

## Solution 2) Sorting then Two Pointer Approach - TC : O(Nlog(n)) , SC : O(N)

Here since we have to return the indices of the pairing numbers hence by just sorting the numbers, will change their indexes values. So in order to preserve the indexes, we need to store the indexes of elements along with their values using a 2D Array.

Then we can sort the array based on the values of the elements, but every element even after sorting will have the index associated with it stored in the array.

Now using two pointer approach we can see if left element and right element both add upto target or not

If their sum> target then we need to reduce the sum which can be done by taking a smaller number ie by reducing the higher number.

Similarly if the sum<target then we need to increase the sum by taking a higher number ie. by increasing the lower number
