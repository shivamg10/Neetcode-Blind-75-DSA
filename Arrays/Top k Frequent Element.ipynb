## Solution 1: Use FrequencyMap to store count of elements and then find Top K frequent Elements in 'K' Iterations    -- TC : O(K * N ) , SC  : O(N)

We can Create a Hashmap in which we'll be storing how many times does a particular element appear in the array ie basically we'll be creating a frequency map. Then We can iterate through all the elements of the freqMap to find the most Frequently occuring element in the array, when we have found that element then we can store that element in the answer array and make sure to remove this element from the map so that it is not chosen again.

## Solution 2) Use FreqMap to store frequency, then sort the elements based on frequency
## TC - O(Nlog(N)), SC: O(N)

Here We can first create a frequency map, then sort the elements based on frequency with most frequenct element at 0th index followed by 2nd most frequent element at 1st index and so on. Finally our answer is the first K elements of that array.

## Solution 3) Using PriorityQueue/heapq  TC - O(Nlog(K)), SC : O(N)

Here we can store the first k element's frequency in priorityQueue/heapq, then whenever another element with more frequency comes up, the element with lowest frequency in the queue is replaced with the new element's frequency.

Since here we need only k elements for comparison and inserting in heapq costs O(log(k)) time, thus for n elements it costs O(N*log(K)) time.

Here we'll be storing the elements corresponding to their frequency in the map, ie here frequency will the acting as key and value will be all those elements which are having that frequency.
