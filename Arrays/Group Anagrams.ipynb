## Solution 1) Sorting every String ,     TC : O(N*Mlog(M)) , SC : O(N) , 
##                                                                          { N - Number of strings , M - size of each String}

Here we will be sorting every string occuring in the list, then that sorted string can be used as key for a hashmap, in which corresponding to a sorted string we will be storing list of valid anagrams

## Solution 2) Use array of size 26 rather than sorting

## TC - O(N *M *26) , SC- O(N * 26)

Here we can create a frequency map of every string and then use it as a key for final hashmap rather than sorting every string
