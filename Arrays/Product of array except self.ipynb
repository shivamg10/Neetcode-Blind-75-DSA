## Solution 2) Taking Whole product then dividing that product by each element
## TC - O(N), SC : O(1)

First we can notice that if the number of zeroes in the array is more than 1 then answer for whole array will be 0 only. Then we can see if we have only 1 zero in the array, then the result of all other array elements will be zero but for the zeroth element it will be product of all other elements. Finally if there is no zero present in the array then we can have product of every element stored in the product variable then divinding that product by the current element will nullify its effect of multiplication in formation of the product.
