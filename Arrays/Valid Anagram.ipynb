## Solution 1 : Sorting - O(Nlog(n))

Firstly compare the length of the strings, bcoz if the lengths are different then we can say directly that the given string are not anagram.

Secondly sort both the strings, by doing so the characters will occurr in the order in both the strings. So we can traverse the string character by character, comparing in both strings.
