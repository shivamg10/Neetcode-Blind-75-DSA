## Solution 1) Brute Force Approach   --- TC : O(N^3*logN), --- SC : O(N) 

Just Search for 3 elements which are having different indexes using 3 for loops and then to remove duplicates, use a set to store these elements.

## Solution 2) Optimised for removing duplicates  --- TC: O(N^3), SC : O(1)

To eliminate duplicates just make sure that same elements don't appear again. To do that sort the list so all the same elements appear consecutively, and then when you select an element check if previous element was same or not. Bcoz if it is same then duplicacy will occurr.

## Solution 3) Two Pointer Approach with a fixed element  -- TC: O(N^2), SC: O(1)

Keep one element fixed then find a pair which has sum equal to -ve of that element or not.
