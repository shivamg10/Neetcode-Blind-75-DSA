## Solution 1) Brute Force --- TC: O(N^2) , SC: O(1)

Find all the possible pairs, then find the area formed by those pair of vertical lines and then find the maximum area.

## Solution 2) Modified Two Pointer Approach  ---TC : O(N), SC: O(1)

Here keep one pointer on the starting tower and the other pointer on the end tower, then find area and update maxArea if required. Then Move the pointer from the side which is having shorter tower. In case when towers are having equal heights then we can move any pointer bcoz you're already bound to at most that height(which is same for both tower) anyways as even if a higher tower comes up then also minimum of two towers height is taken to calculate the area, so the maxArea couldn't be any more than that position anyways since moving either pointer in will yield an area either less than or equal to the area at your current positions.
